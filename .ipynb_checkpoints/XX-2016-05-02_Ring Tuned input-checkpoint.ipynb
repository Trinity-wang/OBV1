# Un anneau pour les gouverner tous...(enfin toutes)

Création d'un réseau récurrent disposant d'une certaine topologie (connexions non aléatoires).
On cherche à ce que les neurones d'une colonne corticale aient une certaine bandwith de selectivité à l'orientation du fait de leurs connexions avec d'autres colonnes.

Pour l'instant, tous les neurones ont le même theta et le w est fixé à 0.
Btheta est donné en degrés

http://neuralensemble.org/docs/PyNN/reference/connectors.html#pyNN.connectors.DistanceDependentProbabilityConnector

http://neuralensemble.org/trac/PyNN/wiki/HighLevelAPI

## defining a tuning function

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=101
angle = 1. * np.arange(N)
```

```python
angle[:, np.newaxis] - angle[np.newaxis, :]
```

```python
tuning_function = lambda i, j, B_theta, N: np.exp( (np.cos(2*(i-j)/N*np.pi)-1.)/B_theta**2)
#def tuning_function(i, j, B_theta, N): return np.exp( (np.cos(2*(i-j)/N*np.pi)-1.)/B_theta**2)
tf = tuning_function(angle[:, np.newaxis], angle[np.newaxis, :], .2, N)
plt.imshow(tf, cmap=plt.gray())
```

```python
tf
```

## setting the weights

```python
import pyNN.nest as sim
```

```python
spike_source = sim.Population(1000, sim.SpikeSourcePoisson(rate=10.))
```

```python
for i, cell in enumerate(spike_source):
    cell.set_parameters(rate=tuning_function(i, 0, .3, 10))
```

```python
def plot_spiketrains(segment):
    fig, ax = plt.subplots(1, 1)
    events, spikes = np.zeros((0,1)), np.zeros((0,1))
    for i, spiketrain in enumerate(segment.spiketrains):
        y = np.array(np.ones_like(spiketrain) * spiketrain.annotations['source_id'])
        try:
            #print((spikes.shape, y.shape))
            spikes = np.concatenate((spikes, y))
            events = np.concatenate((events, i * np.ones_like(y)))
        except: pass
    ax.eventplot(events, spikes)
    plt.ylabel(segment.name)
    plt.setp(plt.gca().get_xticklabels(), visible=False)

plot_spiketrains(net.spikesP)
```

```python
def plot_spiketrains(segment):
    for spiketrain in segment.spiketrains:
        y = np.ones_like(spiketrain) * spiketrain.annotations['source_id']
        plt.plot(spiketrain, y, '.', markersize=.3)
        plt.ylabel(segment.name)
        plt.setp(plt.gca().get_xticklabels(), visible=False)
plot_spiketrains(net.spikesP)
```
